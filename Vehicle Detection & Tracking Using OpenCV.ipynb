# Vehicle Detection & Tracking

- For an Automatic Cars, the detection & Tracking of the Cars Coming from all the Sides is important to make a decision. So, in this project we are going to detect & track the cars coming from 2 directions.

# Project Description

- Here we have used Cascade classifier for Detection & Tracking. Also we have used Cars.xml file which have all the features of the cars included in it.

# 1. Importing Dependencies

```python
# Importing Basic Libraries
import cv2
import imutils
```

```python
cascade_src = 'cars.xml'
car_cascade = cv2.CascadeClassifier(cascade_src)
cam = cv2.VideoCapture('cars.mp4')

while True:
    
    detected = 0
    
    # Reading the frame from camera
    _,img = cam.read()
    
    # Resizing image to 300
    img = imutils.resize(img, width = 300)
    
    # Converting Color to Grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Getting the Coordinates of vehicles in a frame
    cars = car_cascade.detectMultiScale(gray, 1.1, 1) 
    
    for (x,y,w,h) in cars:
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,255), 2)
        
    cv2.imshow("Frame", img)
    
    b = str(len(cars))
    a = int(b)
    
    detected = a
    n = detected
    
    print ("North: %d "%(n))
    
    if n >= 2:
        print ("More Traffic at North.")
    else:
        print ("No Traffic")
    if cv2.waitKey(33) == 27:
        break

cam.release()
cv2.destroyAllWindows()
```

```python
cascade_src = 'cars.xml'
car_cascade = cv2.CascadeClassifier(cascade_src)
cam_1 = cv2.VideoCapture(0)
cam_2 = cv2.VideoCapture('cars.mp4')

# Displaying the scence of 2 Cameras
while True:
    # First Camera Frame
    detected_1 = 0
    
    # Reading the frame from camera
    _, img_1 = cam_1.read()
    
    # Resizing image to 300
    img_1 = imutils.resize(img_1, width = 500)
    
    # Converting Color to Grayscale
    gray_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
    
    # Getting the Coordinates of vehicles in a frame
    cars_1 = car_cascade.detectMultiScale(gray_1, 1.1, 1)
    
    for (x,y,w,h) in cars_1:
        cv2.rectangle(img_1, (x,y), (x+w,y+h), (0,0,255), 2)
        
    cv2.imshow("Car1", img_1)
    
    b_1 = str(len(cars_1))
    a_1 = int(b_1)
    detected_1 = a_1
    n_1 = detected_1
    
    
    # Second Camera Frame
    detected_2 = 0
    
    # Reading the frame from camera
    _, img_2 = cam_2.read()
    
    # Resizing image to 300
    img_2 = imutils.resize(img_2, width = 500)
    
    # Converting Color to Grayscale
    gray_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
    
    # Getting the Coordinates of vehicles in a frame
    cars_2 = car_cascade.detectMultiScale(gray_2, 1.1, 1)
    
    for (x,y,w,h) in cars_2:
        cv2.rectangle(img_2, (x,y), (x+w,y+h), (0,0,255), 2)
        
    cv2.imshow("Camera_2 Scence :", img_2)
    
    b_2 = str(len(cars_2))
    a_2 = int(b_2)
    detected_2 = a_2
    n_2 = detected_2

    
    print ("North: %d "%(n_1))
    if n_1 >= 2:
        print ("More Traffic in North ")
    else:
        print ("No Traffic in North")
        
    print("------------------------------------------------")
    
    print ("South: %d "%(n_2))
    if n_2 >= 2:
        print ("More Traffic in South")
    else:
        print ("No traffic in South")
    
    print("------------------------------------------------")
    print("------------------------------------------------")
    if cv2.waitKey(33) == 27:
        break
        
cam_1.release()    
cam_2.release()
cv2.destroyAllWindows()
```
